```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 300, 301)
y = np.linspace(0, 300, 301)
im = np.meshgrid(x,y)
sigmax, sigmay = 50, 50
amp = 500
background = 10
x0, y0 = 150, 150

X, Y = np.meshgrid(x, y)
fit = amp*np.exp(-((X-x0)**2/(2*sigmax**2) + (Y-y0)**2/(2*sigmay**2)))+background
```

```python
fig, ax = plt.subplots()
pos = ax.imshow(fit)
fig.colorbar(pos)
```
